```python
import requests
import pandas as pd
import numpy as np
import talib

def check_trend_change(prices):
    # Calculate the 50-day and 200-day moving averages
    ma_50 = talib.SMA(prices, timeperiod=50)
    ma_200 = talib.SMA(prices, timeperiod=200)
    print(ma_200)
    if ma_50[-1] > ma_200[-1] and ma_50[-2] <= ma_200[-2]:
        return True  # Trend has changed from bearish to bullish
    elif ma_50[-1] < ma_200[-1] and ma_50[-2] >= ma_200[-2]:
        return True  # Trend has changed from bullish to bearish
    else:
        return False  # No trend change

def analyze_bottom_price(prices, rsi_period=14, rsi_oversold_threshold=30):
    # Calculate the RSI indicator
    rsi = talib.RSI(prices, timeperiod=rsi_period)

    # Check if the current RSI is below the oversold threshold
    if rsi[-1] < rsi_oversold_threshold:
        return True  # Current price is considered the bottom price
    else:
        return False  # Current price is not the bottom price

def check_symbols_kline(symbol, interval, limit):
    url = f'https://contract.mexc.com/api/v1/contract/kline/{symbol}?interval={interval}&limit={limit}'
    #     # time_step = 'Day1' # 合约的参数：间隔: Min1、Min5、Min15、Min30、Min60、Hour4、Hour8、Day1、Week1、Month1，不填时默认Min1

    response = requests.get(url)
    data = response.json()
    df=[]
    try:
        df = pd.DataFrame(data['data'], columns=['time', 'open', 'low', 'high', 'close'])
        df.index = pd.to_datetime(df.time, unit='s', utc=True).map(lambda x: x.tz_convert('Asia/Hong_Kong'))
    except Exception as error:
        print(symbol,'error ', error)

    return df

# API endpoint for historical k-line data

# Parameters for the request
params = {
    "symbol": "btc_usdt",  # Replace with the desired trading pair
    "interval": "1d",      # Replace with the desired time interval (e.g., 1d, 1h, 15m)
    "limit": 1000          # Number of data points to retrieve (max: 1000)
}


df = check_symbols_kline('BTC_USDT', 'Day1', int(200) + 200)



# Convert the closing prices to a numpy array
prices = np.array(df["close"])

# Check if the trend has changed
trend_change = check_trend_change(prices)

# Analyze if the current price is the bottom price
bottom_price = analyze_bottom_price(prices)

print("Trend Change:", trend_change)
print("Bottom Price:", bottom_price)
```

```python

import yfinance as yf
import pandas as pd
import numpy as np
import requests
pd.set_option('display.max_rows', 500)

def get_crypto_list():
    url = "https://api.coingecko.com/api/v3/coins/list"
    response = requests.get(url)
    
    if response.status_code == 200:
        crypto_list = response.json()
        valid_crypto_list = []
        
        for crypto in crypto_list:
            symbol = crypto['symbol'].upper()
            try:
                data = yf.download(symbol + "-USD", period="1d")
                if len(data) > 0:
                    valid_crypto_list.append(symbol)
            except Exception as e:
                print(f"Error processing {symbol}: {str(e)}")
        
        return valid_crypto_list
    else:
        print("Error retrieving crypto list. Status code:", response.status_code)
        return []

def check_ema_cross(crypto_list,check_range):
    end_date = pd.Timestamp.today()
    start_date = end_date - pd.DateOffset(days=check_range+50)

    for crypto in crypto_list:
        try:
            data = yf.download(crypto, start=start_date, end=end_date, progress=False)
            if len(data) >= 60:
                data['ema_10'] = data['Close'].ewm(span=10, adjust=False).mean()
                data['ema_50'] = data['Close'].ewm(span=50, adjust=False).mean()

                ema_10 = data['Close'].ewm(span=10, adjust=False).mean()
                ema_50 = data['Close'].ewm(span=50, adjust=False).mean()
                #data.dropna(inplace=True)

                for i in range(len(data)-check_range,len(data)):
                    if ema_10[i] > ema_50[i] and ema_10[i-1] <= ema_50[i-1]:
                        cross_date = data.index[i]
                        print(f"{crypto}: 10 EMA crossed above 50 EMA on {cross_date}")
        except:
            print('Cannot download ', crypto)
# Example usage

crypto_list = ['BTC-USD', 'ETH-USD','LINK-USD']
crypto_list = get_crypto_list()
check_ema_cross(crypto_list,30)
```
